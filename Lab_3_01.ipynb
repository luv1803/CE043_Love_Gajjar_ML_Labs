```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.naive_bayes import MultinomialNB,GaussianNB
```

```python
data = pd.read_csv('/content/Dataset3.csv')
data
```

```python
le = preprocessing.LabelEncoder()
outlook_encoded = le.fit_transform(data['Outlook'])
print("Outlook_Encoded::",outlook_encoded)
```

```python
temp_encoded = le.fit_transform(data['Temp'])
print("Temp_Encoded::",temp_encoded)
```

```python
humidity_encoded = le.fit_transform(data['Humidity'])
print("Humdity_Encoded::",humidity_encoded)
```

```python
wind_encoded = le.fit_transform(data['Wind'])
print("Wind_Encoded::",wind_encoded)
```

```python
features = tuple(zip(outlook_encoded,temp_encoded,wind_encoded,humidity_encoded))
print("FEATURES::",features)
```

```python
labels = le.fit_transform(data['Class'])
print("LABELS::",labels)
```

```python
from sklearn.model_selection import train_test_split
train_dataX,test_dataX = train_test_split(features,test_size = 0.05,random_state = 10)
print("TRAINING EXAMPLES::",train_dataX)
print("TESTING EXAMPLES::",test_dataX)

trainY,testY = train_test_split(labels,test_size = 0.05,random_state = 10)
print("TRAIN_Y::",trainY)
print("TEST_Y::",testY)
```

```python
#Creating a classifier
model = MultinomialNB()
#training the classifier
model.fit(train_dataX,trainY)
```

```python
predicted= model.predict([[2,2,0,2]]) 
print("Predicted Value for :", predicted)
```
