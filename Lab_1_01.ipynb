```python
import numpy as np
import pandas as pd
import re
import string
import math
data = pd.read_csv('/content/sample_data/emails.csv', usecols=['text','spam'])
data.rename(columns={'spam':'class'},inplace=True)
data['label'] = np.where(data['class']==1,'spam','ham')
data.drop_duplicates(inplace=True)
print(data)
print(data['text'][0])
```

```python
#lower case
data['text'][0] = data['text'][0].lower()
print(data['text'][0])
```

```python
#remove punctuation
data['text'] = data['text'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
print(data['text'][0])
```

```python
#remove stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words.add('subject')
stop_words.add('http')
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
data['text'] = data['text'].apply(lambda x: remove_stopwords(x))
print(data['text'])
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
data["text"] = data["text"].apply(lambda x: stem_words(x))
print(data['text'])
```
